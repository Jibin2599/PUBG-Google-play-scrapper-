```python
from google_play_scraper import Sort,reviews
result,continuation_token=reviews(
'com.pubg.imobile',
lang='en',
country='us',
sort=Sort.NEWEST,
count=1000,
filter_score_with=5)
result,_=reviews('com.pubg.imobile',
                continuation_token=continuation_token)
```

```python
print(result)
```

```python
import pandas as pd
import numpy as np
```

```python
pubg=pd.DataFrame(np.array(result),columns=['review'])
pubg=pubg.join(pd.DataFrame(pubg.pop('review').tolist()))
pubg.head()
```

```python
from IPython.display import HTML
import base64
def create_download_link( data, title = "Download CSV file", filename = "pubg.csv"):  
    csv = pubg.to_csv()
    b64 = base64.b64encode(csv.encode())
    payload = b64.decode()
    html = '<a download="{filename}" href="data:text/csv;base64,{payload}" target="_blank">{title}</a>'
    html = html.format(payload=payload,title=title,filename=filename)
    return HTML(html)

create_download_link(pubg)
```
